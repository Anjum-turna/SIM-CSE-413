```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Generate 100,000 random numbers in [0,1)
random_numbers = np.random.rand(10)

# Plot the histogram
plt.hist(random_numbers, bins=5,  alpha=0.75, color='blue', edgecolor='white')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Uniformly Distributed Random Numbers')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns


# Generate 10,000 random values from a normal distribution
SNormalValues = np.random.randn(10000)#mean=0, standard deviation=1

# Create a histogram
plt.hist(SNormalValues, bins=30, edgecolor='black')
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Generate 10,000 random values from a normal distribution with mean=100 and std=10
NormalValues = np.random.normal(0, 1, 500)

# Create a histogram
plt.hist(NormalValues, bins=30, edgecolor='black')
plt.title('Histogram of Normal Distribution (mean=0, std=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# With Seed
np.random.seed(10)  # Setting seed to 10
seeded_matrix = np.random.randint(1, 51, (3, 3))
print("Seeded Random Matrix:\n", seeded_matrix)

```

as i set the seed to a positive integer in 10 so the numbers between 1 to 50 values doesnt changes as per as i run the cell.It is fixed.

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Without Seed
np.random.seed(None)  # No seed, generates different values each time
non_seeded_matrix = np.random.randint(1, 51, (3, 3))
print("\nNon-Seeded Random Matrix:\n", non_seeded_matrix)
```

as i set the seed to none so between 1 to 50 value changes as per as i run the cell.
