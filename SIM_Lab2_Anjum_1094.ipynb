```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Assignment 1

# Create two random 3x3 matrices with integer values between 0 and 10
A = np.random.randint(0, 10, size=(3, 3))  # Random integers between 0 and 9
B = np.random.randint(0, 10, size=(3, 3))  # Random integers between 0 and 9

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Determinant of a matrix
det_A = np.linalg.det(A)  # Returns the determinant of the matrix

# Inverse of a matrix
if np.linalg.det(A) != 0:  # Check if matrix is invertible
    inv_A = np.linalg.inv(A)  # Compute the inverse of A
else:
    inv_A = "Matrix is singular, cannot compute inverse"  # No inverse if determinant is 0

# Rank of a matrix
rank_A = np.linalg.matrix_rank(A)  # Returns the rank of the matrix

# Matrix operations
matrix_sum = A + B  # Matrix addition
matrix_diff = A - B  # Matrix subtraction
matrix_prod = np.dot(A, B)  # Matrix multiplication (dot product)

# Display results
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Addition: ", matrix_sum)
print("Subtractiion: ", matrix_diff)
print("Multiplication: ", matrix_prod)
print("Determinant of A:", det_A)
print("Inverse of A:\n", inv_A)
print("Rank of A:", rank_A)

```

```python
# Generate two vectors with 15 random floats
vector1 = np.random.rand(15)
vector2 = np.random.rand(15)

# Create an index for x-axis
x = np.arange(15)

# Plot the vectors
plt.plot(x, vector1, marker='o', linestyle='-', label='Vector 1')
plt.plot(x, vector2, marker='s', linestyle='--', label='Vector 2')

# Labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Comparison of Two Random Vectors')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Generate a 4x4 matrix with random values
matrix = np.random.rand(4, 4)

# Create a heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(matrix, annot=True, cmap='viridis', xticklabels=np.arange(1, 5), yticklabels=np.arange(1, 5))

# Labels and title
plt.xlabel('Column')
plt.ylabel('Row')
plt.title('Heatmap of 4x4 Random Matrix')

# Show the plot
plt.show()

# Create two 4x4 matrices with random integer values between 0 and 10
A = np.random.randint(0, 10, size=(4, 4))
B = np.random.randint(0, 10, size=(4, 4))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Matrix operations
matrix_sum = A + B  # Matrix addition
matrix_diff = A - B  # Matrix subtraction
matrix_prod = np.dot(A, B)  # Matrix multiplication

# Flatten matrices for bar plotting
sum_flat = matrix_sum.flatten()
diff_flat = matrix_diff.flatten()
prod_flat = matrix_prod.flatten()
x = np.arange(len(sum_flat))

# Create bar plots for each operation
plt.figure(figsize=(15, 5))

# Bar plot for matrix addition
plt.subplot(1, 3, 1)
plt.bar(x, sum_flat, color='skyblue')
plt.xlabel('Element Index')
plt.ylabel('Value')
plt.title('Matrix Addition (A + B)')

# Bar plot for matrix subtraction
plt.subplot(1, 3, 2)
plt.bar(x, diff_flat, color='lightcoral')
plt.xlabel('Element Index')
plt.ylabel('Value')
plt.title('Matrix Subtraction (A - B)')

# Bar plot for matrix multiplication
plt.subplot(1, 3, 3)
plt.bar(x, prod_flat, color='lightgreen')
plt.xlabel('Element Index')
plt.ylabel('Value')
plt.title('Matrix Multiplication (A  B)')

plt.tight_layout()
plt.show()
```
